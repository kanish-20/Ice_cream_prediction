```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


```

```python
df=pd.read_csv('/content/Ice Cream.csv')
df.head()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='Temperature', y='Revenue', data=df)
plt.title('Temperature vs Revenue')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.show()


print("Correlation:", df['Temperature'].corr(df['Revenue']))
```

```python
X=df[['Temperature']]
Y=df['Revenue']
```

```python
X_train = np.array('Temperature').reshape(-1, 1)
Y_train = np.array('Revenue')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Revenue=model.predict(X_test)
```

```python
print("MSE:",mean_squared_error(Y_test,Revenue))
```

```python
r2 = r2_score(Y_test, Revenue)
print("Accuracy:", r2)
```
